```python
book = book_df[['ISBN', 'Book-Title', 'Book-Author', 'Year-Of-Publication', 'Publisher','Image-URL-L']]
book.head()
```

```python
book.rename(columns = {"ISBN" : "ISBN",
                    "Book-Title" : "Title",
                    "Book-Author" : "Author",
                    "Year-Of-Publication" : "Year",
                    "Publisher" : "Publisher",
                    "Image-URL-L" : "Image"}, inplace=True)

print(f"shape of dataset : {book.shape}")
book.head()
df = book[book['Title']=='2nd Chance']
df.head()
```

```python
user_df = pd.read_csv("dataset\BX-Users.csv", sep=";", encoding="latin1", on_bad_lines="skip")
print(f"column of dataset : {user_df.columns}")
print(f"shape of dataset : {user_df.shape}")

user_df.head()
```

```python
user = user_df.rename(columns={"User-ID" : "Id"})
user.head()
```

```python
rating_df = pd.read_csv("dataset\BX-Book-Ratings.csv", sep=";", encoding="latin1", on_bad_lines="skip")
print(f"shape of dataset : {rating_df.shape}")
print(f"column of dataset : {rating_df.columns}")
rating_df.head()
```

```python
rating = rating_df.rename(columns={"User-ID" : "Id", 
                                   "Book-Rating" : "Rating"})
rating.head()
```

```python
rating.columns, user.columns, book.columns
```

```python
count_user_rate = rating['Id'].value_counts()
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
axes[0].boxplot(count_user_rate)
axes[0].set_title("Boxplot")
axes[0].set_ylabel("Values")

# Histogram
axes[1].hist(count_user_rate, bins=100, color='skyblue', edgecolor='black')
axes[1].set_title("Histogram")
axes[1].set_xlabel("Bins")
axes[1].set_ylabel("Frequency")
sum = 0
counts, bins, patches = plt.hist(count_user_rate, bins=100, edgecolor="black", alpha=0)
print("Bin Ranges and Counts:")
for i in range(len(bins)-1):
    #print(f"Bin {i+1}: {bins[i]:.1f} - {bins[i+1]:.1f}, Count: {int(counts[i])}")
    sum += int(counts[i])
print("-"*50)
print(f"shape of rating dataset (before) : {rating.shape}")
print(f"Id of rating dataset (before) : {len(rating['Id'].unique())}")
plt.tight_layout()
plt.show()
```

```python
# create series โดย id เป็น index และค่าใน series คือจำนวนครั้งการคะแนนของแต่ละ id (rating['Id'].value_counts())
## หากมี condition เกิดขึ้น จะแปลงตัวเลขจาก series ให้เป็น boolen ว่ามีการให้ตะแนนมากกว่า 150 ครั้งหรือไม่ (rating['Id'].value_counts() > 150)

count_user_rate = rating['Id'].value_counts() > 150
count_user_rate_True = count_user_rate[count_user_rate == True]
rating = rating[rating["Id"].isin(count_user_rate_True.index)]

fig, axes = plt.subplots(1, 2, figsize=(12, 5))
axes[0].boxplot(rating['Id'].value_counts())
axes[0].set_title("Boxplot")
axes[0].set_ylabel("Values")

# Histogram
axes[1].hist(rating['Id'].value_counts(), bins=100, color='skyblue', edgecolor='black')
axes[1].set_title("Histogram")
axes[1].set_xlabel("Bins")
axes[1].set_ylabel("Frequency")
sum = 0
counts, bins, patches = plt.hist(rating['Id'].value_counts(), bins=100, edgecolor="black", alpha=0)
print("Bin Ranges and Counts:")
for i in range(len(bins)-1):
    #print(f"Bin {i+1}: {bins[i]:.1f} - {bins[i+1]:.1f}, Count: {int(counts[i])}")
    sum += int(counts[i])
print("-"*50)
print(f"shape of rating dataset (before) : {rating.shape}")
print(f"Id of rating dataset (before) : {len(rating['Id'].unique())}")
plt.tight_layout()
plt.show()
```

```python
rating_with_book = rating.merge(book, on='ISBN')
print(f"rating-with-book : {rating_with_book.shape}")
rating_with_book.head()
```

```python
# number rating หนังสือถูก rating ไปทั้งหมดกี่ครั้ง
number_rating = rating_with_book.groupby('Title')['Rating'].count().reset_index()
number_rating = number_rating.rename(columns={'Rating' : 'Num_of_rating'})
number_rating.head()
```

```python
final_rating = rating_with_book.merge(number_rating, on='Title')
final_rating.head()
```

```python
# ตรวจสอบค่าของ max_final_rating
max_final_rating = final_rating['Num_of_rating']
print(max_final_rating.head())  # แสดงข้อมูล 5 แถวแรก

# สร้าง Boxplot และ Histogram
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Boxplot
axes[0].boxplot(max_final_rating)
axes[0].set_title("Boxplot")
axes[0].set_ylabel("Values")

# Histogram
axes[1].hist(max_final_rating, bins=100, color='skyblue', edgecolor='black')
axes[1].set_title("Histogram")
axes[1].set_xlabel("Bins")
axes[1].set_ylabel("Frequency")

# แสดงกราฟ
plt.tight_layout()
plt.show()
```

```python
final_rating = final_rating[final_rating['Num_of_rating'] >= 50]
final_rating
```

```python
final_rating.drop_duplicates(['Id', 'Title'], inplace=True)
final_rating
```

```python
book_pivot = final_rating.pivot_table(columns='Id', index='Title', values='Rating')
book_pivot
```

```python
book_pivot.fillna(0, inplace=True)
book_pivot
```

```python
from scipy.sparse import csr_matrix
book_sparse = csr_matrix(book_pivot)
print(book_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors
knn_model = NearestNeighbors()
knn_model.fit(book_sparse)
```

```python
book_pivot.iloc[0, :].values.reshape(1, -1).shape
```

```python
distances, indices = knn_model.kneighbors(book_pivot.iloc[0, :].values.reshape(1, -1), n_neighbors=6)
print("Indices of nearest neighbors:", indices)
print("Distances to nearest neighbors:", distances)
```

```python
distances, indices = knn_model.kneighbors(book_pivot.iloc[0, :].values.reshape(1, -1), n_neighbors=6)
for i in indices:
    print(book_pivot.index[i])
```

```python
# function 
def recommend_book(name):
    book_idx = book_pivot.index.get_loc(name)
    dist, idx = knn_model.kneighbors(book_pivot.iloc[book_idx, :].values.reshape(1, -1), n_neighbors=5)
    # print(idx) # -> [[  0 384 434 270 808]]
    for i in range (1, len(idx[0])):
        idx_book = idx[0][i]
        print(idx_book)
        print(book_pivot.index[int(idx_book)])
        print(book[book['Title']==book_pivot.index[int(idx_book)]]['Image'].values[0])
        print("-"*30)
recommend_book('2nd Chance')
```
